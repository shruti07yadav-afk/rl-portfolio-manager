TRADING ENVIROMENT

```python
class TradingEnvironment:

    def __init__(self, data):
        self.data = data
        self.current_step = 0
        self.cash = 100000
        self.shares = 0

    def reset(self):
        self.current_step = 0
        self.cash = 100000
        self.shares = 0

        return self.data.iloc[self.current_step]

    def step(self, action):

        current_price = self.data.iloc[self.current_step]['Close']

        portfolio_before = self.cash + self.shares * current_price

        if action == 1 and self.cash >= current_price:
            self.shares += 1
            self.cash -= current_price

        elif action == 2 and self.shares > 0:
            self.shares -= 1
            self.cash += current_price

        self.current_step += 1

        done = self.current_step >= len(self.data) - 1

        next_price = self.data.iloc[self.current_step]['Close']

        portfolio_after = self.cash + self.shares * next_price

        reward = portfolio_after - portfolio_before

        next_state = self.data.iloc[self.current_step]

        return next_state, reward, done
```

PROTFOLIO MANAGEMENT

```python
class PortfolioEnvironment:

    def __init__(self, data):

        self.data = data
        self.current_step = 0

        self.cash = 100000

        self.itc_shares = 0
        self.hdfc_shares = 0
        self.tcs_shares = 0

    def reset(self):

        self.current_step = 0

        self.cash = 100000

        self.itc_shares = 0
        self.hdfc_shares = 0
        self.tcs_shares = 0

        return self.data.iloc[self.current_step]

    def get_portfolio_value(self):

        row = self.data.iloc[self.current_step]

        return (
            self.cash
            + self.itc_shares * row['ITC_Close']
            + self.hdfc_shares * row['HDFC_Close']
            + self.tcs_shares * row['TCS_Close']
        )

    def step(self, action):

        row = self.data.iloc[self.current_step]

        itc_price = row['ITC_Close']
        hdfc_price = row['HDFC_Close']
        tcs_price = row['TCS_Close']

        portfolio_before = self.get_portfolio_value()

        # ITC Actions
        if action == 1 and self.cash >= itc_price:
            self.itc_shares += 1
            self.cash -= itc_price

        elif action == 2 and self.itc_shares > 0:
            self.itc_shares -= 1
            self.cash += itc_price

        # HDFC Actions
        elif action == 3 and self.cash >= hdfc_price:
            self.hdfc_shares += 1
            self.cash -= hdfc_price

        elif action == 4 and self.hdfc_shares > 0:
            self.hdfc_shares -= 1
            self.cash += hdfc_price

        # TCS Actions
        elif action == 5 and self.cash >= tcs_price:
            self.tcs_shares += 1
            self.cash -= tcs_price

        elif action == 6 and self.tcs_shares > 0:
            self.tcs_shares -= 1
            self.cash += tcs_price

        self.current_step += 1

        done = self.current_step >= len(self.data) - 1

        portfolio_after = self.get_portfolio_value()

        # UPDATED REWARD FUNCTION (percentage return)
        reward = (
            portfolio_after - portfolio_before
        ) / portfolio_before

        next_state = self.data.iloc[self.current_step]

        return next_state, reward, done
```

```python
import gymnasium as gym
from gymnasium import spaces

import numpy as np

from stable_baselines3 import PPO
```

```python
class PortfolioGymEnvironment(gym.Env):

    def __init__(self, data):

        super().__init__()

        self.env = PortfolioEnvironment(data)

        self.action_space = spaces.Discrete(7)

        self.observation_space = spaces.Box(
            low=-np.inf,
            high=np.inf,
            shape=(18,),
            dtype=np.float32
        )

    def reset(self, seed=None, options=None):

        state = self.env.reset()

        return state.values.astype(np.float32), {}

    def step(self, action):

        next_state, reward, done = self.env.step(action)

        terminated = done
        truncated = False

        return (
            next_state.values.astype(np.float32),
            reward,
            terminated,
            truncated,
            {}
        )
```

```python
gym_env = PortfolioGymEnvironment(portfolio_features)

state, info = gym_env.reset()

print(state.shape)
```

```python
next_state, reward, terminated, truncated, info = gym_env.step(1)

print(reward)
print(terminated)
```

```python
model = PPO(
    "MlpPolicy",
    gym_env,
    verbose=1
)
```

```python
model.learn(total_timesteps=10000)
```

```python
obs, _ = gym_env.reset()

done = False

total_reward = 0

while not done:

    action, _ = model.predict(obs)

    obs, reward, terminated, truncated, info = gym_env.step(action)

    total_reward += reward

    done = terminated or truncated

print("Total Reward:", total_reward)
```
